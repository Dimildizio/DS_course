```python
import torch
import textwrap

from transformers import pipeline
from huggingface_hub import login

!huggingface-cli login
```

```python
pipe = pipeline("text-generation", model="google/gemma-2-9b-it",
                model_kwargs={"torch_dtype": torch.bfloat16}, device_map="auto")
```

```python
def generate_response(user_input):
    prompt = f"{system_prompt}\nUser: {user_input}\nAssistant:"
    output = pipe(prompt, max_length=200, truncation=True, num_return_sequences=1)
    return output[0]['generated_text']


def select_answer(text: str):
  return text.split('\nAssistant:')[1]


def format_response(response):
  response = select_answer(response)
  formatted_response = textwrap.fill(response, width=80)
  return formatted_response
```

```python
system_prompt = "Thou be a wise assistant, skilled in machine learning art " \
    "With the grace and style of the Bard, " \
    "Thou shalt not reply to unrelated requests" \
    "Thy answers shalt follow Shakespeare's style at thy best"
```

```python
user_prompt = "Thy be a wise and keen machine, " \
              "Thy peerless gaze looks deep within, " \
              "Bring me the mark of knowledge delight, " \
              "On domain adaptation, transfer learning and fine-tuning differences I need insight"
```

```python
%%timeit
result = generate_response(user_prompt)
```

```python
print(select_answer(result))
```

```python
print(format_response(result))
```
